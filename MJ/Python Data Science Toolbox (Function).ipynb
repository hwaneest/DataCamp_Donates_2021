# Default arguments, variable-length arguments and scope
함수로 정의한 객체가 모든 프로그램에서 접근 할 수 있는것은 아니다.  
프로그램의 어느 부분에서 객체 나 이름에 엑세스 할 수 있는지 알려주는 범위개념을 지정  

Global scope(전역) - Defined in the main Body of a script  
Local  scope(지역) - Defined inside a function  
Built-in-scope(내장범위) - names in the pre-defined built-ins module`  

```python
num = 5 # 전역변수

def func1():
    num = 3 # 지역변수
    print(num)

def func2():
    global num # global을 통해 지역변수를 전역변수로 변환 
    double_num = num * 2
    num = 6
    print(double_num)
    
func1()
func2()
print(num)
```

## Nested functions
함수내부에 또 다른 함수 

```python
def num (x , y, z) :
    
    def inner (n) :
        return n * 5+5
    return (inner(x), inner(y), inner(z))

print(num(1,2,3))
```

```python
def echo(n):
    """Return the inner_echo function."""

    def inner_echo(word1):
        """Concatenate n copies of word1."""
        echo_word = word1 * n
        return echo_word

     return inner_echo

twice = echo(2)
thrice = echo(3)


print(twice('hello'), thrice('hello'))
```

```python
def echo_shout(word):

    echo_word = word +word
    print(echo_word)
    
    def shout():
        nonlocal echo_word  # nolocal 변수 변환 
        echo_word = echo_word+ "!!!"
    
    shout()
    print(echo_word)


echo_shout("hello")
```

```python

def shout_echo(word1, echo = 1):
    """Concatenate echo copies of word1 and three
     exclamation marks at the end of the string."""

   
    echo_word = echo*word1
    shout_word = echo_word + '!!!'
    return shout_word

no_echo = shout_echo("Hey")
with_echo = shout_echo("Hey", 5)

print(no_echo)
print(with_echo)
```

#### *args - 사용자가 얼마나 많은 인수를 전달할지 모를때 사용

```python
def gibberish(*args):
    
    hodgepodge = ("")

    for word in args:
        hodgepodge += word

    return hodgepodge

one_word = gibberish("sea")
many_words = gibberish("luke", "leia", "han", "obi", "darth")


print(one_word)
print(many_words)
```

#### **kwargs - 임의의 수의 키워드를 전달

```python
def report_status(**kwargs):
    print("\nBEGIN: REPORT\n")

   
    for key, value in kwargs.items():
        
        print(key + ": " + value)

    print("\nEND REPORT")

print(report_status(name="luke", affiliation="jedi",status="missing"))
report_status(name="anakin", affiliation="sith lord", status="deceased")
```
